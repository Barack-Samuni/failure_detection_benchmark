```python
import os
import sys
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

root = "/data/failure_detection"
sys.path.append(root)
from sklearn.manifold import TSNE
from configs.load_model_and_config import get_config_data_model_for_eval
from failure_detection.gather_results_utils import gather_single_pass_outputs

os.environ["OUTPUT_DIR"] = "/data/failure_detection/outputs"
figure_dir = Path("/data/failure_detection/outputs/figures")
```

```python
config, data_module, model, output_dir = get_config_data_model_for_eval(
    "medmnist/tissuemnist_resnet_dropout_all_layers.yml"
)
model = model[0].cuda()
train_out = gather_single_pass_outputs(
    model, data_module.train_dataloader(), config.n_classes
)
tsne = TSNE(n_components=2, random_state=0)
X_2d = tsne.fit_transform(train_out["feats"].numpy())
plt.figure(figsize=(5, 5))
sns.set_theme(style="dark", palette="colorblind")
indices_to_plot = np.random.choice(X_2d.shape[0], np.min([X_2d.shape[0], 5000]))
sns.scatterplot(
    x=X_2d[indices_to_plot, 0],
    y=X_2d[indices_to_plot, 1],
    hue=train_out["targets"].int().numpy().astype(str)[indices_to_plot],
    alpha=0.5,
    hue_order=np.arange(config.n_classes).astype(str),
)
plt.xticks([])
plt.yticks([])
plt.savefig(fname=str(figure_dir / "tsne_tissuemnist.pdf"), bbox_inches="tight")
```
